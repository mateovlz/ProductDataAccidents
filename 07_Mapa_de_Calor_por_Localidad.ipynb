```python

from matplotlib.path import Path
import matplotlib.pyplot as plt
import numpy as np

import cartopy.crs as ccrs
import cartopy.feature as cf
import cartopy
from cartopy.io.img_tiles import OSM

from matplotlib.transforms import offset_copy

import cartopy.io.img_tiles as cimgt


```

```python

def main():
    imagery = OSM()
    plt.figure(figsize = (18, 20))
    ax = plt.axes(projection=imagery.crs)
    ax.set_extent((-74.400, -73.8, 4, 5))

    
    def point(x , y , nombre, num, color):
        ax.plot(x, y, marker='o', color=color, markersize=0.003*num,
            alpha=0.7, transform=ccrs.Geodetic())
        plt.text(x, y, nombre,
            horizontalalignment = 'right',
            transform = ccrs.Geodetic())
        plt.text(x-0.02, y+0.0085, num,
            horizontalalignment = 'left',
            transform = ccrs.Geodetic())

    # Add the imagery to the map.
    
    ax.add_image(imagery, 15)
    ax.add_feature(cf.STATES)

    # Add the locations, first with the red circlw,
    #Barrios unidos
  
    point(-74.084 ,  4.666 , 'Barrios Unidos', 11329,'blue')
    point(-74.107 ,  4.707 , 'Engativa', 20469, 'green')
    point(-74.315 ,  4.035 , 'Sumapaz',6,'red')
    point(-74.094 ,  4.645 , 'Teusaquillo', 11129,'cyan' )
    point(-74.074 ,  4.594 , 'La Candelaria', 915, 'magenta')
    point(-74.030 ,  4.596 , 'Santa Fe', 5781, 'yellow')
    point(-74.082 ,  4.765 , 'Suba', 17072, 'blue')
    point(-74.148 ,  4.683 , 'Fontibón', 16989, 'green')
    point(-74.091 ,  4.603 , 'Los Martires', 6820, 'red')
    point(-74.088 ,  4.546 , 'San Critobal', 4984, 'cyan')
    point(-74.103 ,  4.477 , 'Usme', 3536, 'magenta')
    point(-74.123 ,  4.615 , 'Puente Aranda', 14063, 'yellow')
    point(-74.031 ,  4.749 , 'Usaquén', 21319, 'blue')
    point(-74.195 ,  4.631 , 'Bosa', 9764, 'green')
    point(-74.154 ,  4.507 , 'Ciudad Bolivar', 7010, 'red')
    point(-74.116 ,  4.565 , 'Rafael Uribe Uribe', 4483, 'cyan')
    point(-74.157 ,  4.627 , 'Kennedy', 22127, 'magenta')
    point(-74.047 ,  4.657 , 'Chapinero', 11806, 'yellow')
    point(-74.141 ,  4.588 , 'Tunjuelito', 5309, 'blue')
    point(-74.101 ,  4.549 , 'Antonio Nariño', 3830, 'green')

    plt.title('Siniestros por localidad')
    plt.show()
    plt.close()
    plt.savefig('mapa_localidades')

    
if __name__ == '__main__':
    main()
```
