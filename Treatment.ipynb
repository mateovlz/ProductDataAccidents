# ACCIDENTES 2017

```python
df2017AC = pd.read_excel(file2017)
```

#Se convierten nombres de columnas a mayusculas y se despliegan columnas.

```python
for column in df2017AC.columns:
    df2017AC.rename(columns = {column:column.upper()}, inplace = True)
print(df2017AC.columns.values)
```

```python
df2017AC.head()
```

# Analisis de datos Accidentes 2017

```python
df2017AC.describe()
```

Se addiciona columna KeyId para diferenciar datasets por Año

```python
df17AC = df2017AC
lskeys = []
for index, row in df2017AC.iterrows():
    fecha = row['FECHA']
    fechaDate = fecha.to_pydatetime()
    keyid = str(row['ID']) + str(fechaDate.day) + str(fechaDate.month) + str(fechaDate.year)
    lskeys.append(int(keyid))
    
df17AC.insert(0,"KEYID", lskeys)
```
